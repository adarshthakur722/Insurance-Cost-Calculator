```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
smoker_value = {'no': 0, 'yes': 1}
df['Smoking'] = df.smoker.map(smoker_value)
```

```python
df.drop(columns=['smoker'], axis = 1, inplace = True)
```

```python
df
```

```python
encoded = ohe.fit_transform(df[['sex']]).toarray()
```

```python
ohe.get_feature_names_out()
```

```python
new_df = pd.DataFrame(encoded, columns = ohe.get_feature_names_out(), dtype=int)
```

```python
df.drop(columns=['sex'], axis = 1, inplace=True)
```

```python
new_df = pd.concat([df,new_df], axis = 1)
```

```python
new_df
```

```python
encoded = ohe.fit_transform(df[['region']]).toarray()
```

```python
region_df = pd.DataFrame(encoded, columns = ohe.get_feature_names_out(), dtype=int)
```

```python
region_df
```

```python
new_df = pd.concat([new_df,region_df], axis = 1)
```

```python
new_df.drop(columns=['region'], axis = 1, inplace = True)
```

```python
new_df.sample(10)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV 
```

```python
rf = RandomForestRegressor()
lr = LinearRegression()
dt = DecisionTreeRegressor()
svr = SVR()
```

```python
y=new_df['charges']
X = new_df.drop(columns = ['charges'], axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=2)
```

```python
estimators = [('lr', lr), ('dt', dt), ('svr', svr), ('rf', rf)]
```

```python
for estimator in estimators:
    x = cross_val_score(estimator[1], X,y, cv = 15, scoring = 'r2')
    print(estimator[0], np.round(np.mean(x),2))
```

```python
X_test
```

```python
y_test
```

```python
parameters = {
    'n_estimators': [50, 100, 150, 200],          
    'max_depth': [None, 10, 20, 30, 40],                           
    'max_features': ['auto', 'sqrt', 'log2'],                         
    'criterion': ['squared_error', 'absolute_error', 'friedman_mse', 'poisson']                    
}
```

```python
from sklearn.metrics import r2_score
```

```python
grid = GridSearchCV(rf, param_grid=parameters, cv=5,verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
model = RandomForestRegressor(criterion='absolute_error', max_depth= 10, max_features = 'sqrt', n_estimators = 150)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## R2 Score of the Model

```python
 r2_score(y_test,y_pred)
```
